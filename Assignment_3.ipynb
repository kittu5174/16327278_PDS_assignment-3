```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

data = pd.read_csv('Corona_NLP_test.csv')
tweets = data['OriginalTweet']
tokens = []
for tweet in tweets:
    tweet = tweet.lower()
    tweet_tokens = nltk.word_tokenize(tweet)
    tokens.extend(tweet_tokens)

stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words]
word_counts = Counter(filtered_tokens)
wordcloud = WordCloud(width=800, height=800, background_color='white', colormap='inferno').generate_from_frequencies(word_counts)
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
